```python
from transformers import pipeline
from langchain_community.llms import HuggingFacePipeline
```

```python
pipe = pipeline("text-generation", model="gpt2")

llm = HuggingFacePipeline(pipeline=pipe)
```

```python
print(llm.invoke("Explain Artificial Intelligence in simple terms"))
```

```python
# 🔹 Simple Tool Function (Calculator)
def calculator(query):
    try:
        return eval(query)
    except:
        return "Invalid input"

# Example usage
question = "15 + 25"

print("User Question:", question)
print("Tool Output:", calculator(question))
```

```python
# Simple LLM call
print("🔹 LLM Example:")
print(llm.invoke("Explain Artificial Intelligence in simple terms"))
```

```python
from transformers import pipeline
from langchain_community.llms import HuggingFacePipeline

from langchain_core.prompts import PromptTemplate
from langchain_core.runnables import RunnableSequence   # ✅ replaces LLMChain
```

```python
print("\n🔹 Prompt Template Example:")

prompt = PromptTemplate.from_template("Explain {topic} in simple terms")

formatted_prompt = prompt.format(topic="Machine Learning")
print(formatted_prompt)

print(llm.invoke(formatted_prompt))
```

```python
print("\n🔹 Chain Example:")

chain = prompt | llm   # ✅ new pipeline style

result = chain.invoke({"topic": "Deep Learning"})
print(result)
```

```python
print("\n🔹 Memory Example:")

chat_history = []

chat_history.append(("Hello", "Hi there!"))
chat_history.append(("What is AI?", "AI is Artificial Intelligence"))

for q, a in chat_history:
    print("User:", q)
    print("Bot:", a)
```

```python
print("\n🔹 Tool Example:")

def calculator(query):
    try:
        return eval(query)
    except:
        return "Invalid input"

question = "15 + 25"

print("User Question:", question)
print("Tool Output:", calculator(question))
```
